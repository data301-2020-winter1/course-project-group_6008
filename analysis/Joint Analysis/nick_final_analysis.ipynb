### Research Question: Which US University Produced the Best Players in Each Position in the 2014-2015 Season?

```python
import pandas as pd
import pandas_profiling
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Step #1 - Load our data and remove any n/a values

```python
players_stats_og = (
pd.read_csv('/Users/Nick/Desktop/Final Project/data/raw/players_stats.csv')
.dropna()
.reset_index()
.drop(['index'], axis = 1)
.rename(columns={"Collage" : "College", "Pos" : "Position"})
)

players_stats_og
```

```python
players_stats_university = (
        players_stats_og[players_stats_og['College'].notnull()]
        .reset_index()
        .drop(columns=['index','Age','Games Played', 'MIN', 'Birthdate','Birth_Place', 'Experience', 'Height', 'Team', 'Weight', 'BMI']) 
        )
```

```python
players_stats_PG = (
        players_stats_university.loc[players_stats_university['Position'] == 'PG']
        .sort_values(by = 'EFF', ascending=False)
        .reset_index()
        .drop(columns=['index'])
        )
    
players_stats_SG = (
        players_stats_university.loc[players_stats_university['Position'] == 'SG']
        .sort_values(by = 'EFF', ascending=False)
        .reset_index()
        .drop(columns=['index'])
        )

players_stats_SF = (
        players_stats_university.loc[players_stats_university['Position'] == 'SF']
        .sort_values(by = 'EFF', ascending=False)
        .reset_index()
        .drop(columns=['index'])
        )

players_stats_PF = (
        players_stats_university.loc[players_stats_university['Position'] == 'PF']
        .sort_values(by = 'EFF', ascending=False)
        .reset_index()
        .drop(columns=['index'])
        )

players_stats_C = (
        players_stats_university.loc[players_stats_university['Position'] == 'C']
        .sort_values(by = 'EFF', ascending=False)
        .reset_index()
        .drop(columns=['index'])
        )
```
