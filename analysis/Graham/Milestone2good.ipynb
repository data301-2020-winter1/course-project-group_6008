```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
```

```python
df = pd.read_csv('../../data/raw/players_stats.csv')
def load_and_clean(url_or_path_to_csv_file):
    df = pd.read_csv(url_or_path_to_csv_file)
    df1 = df.drop(columns = ['BMI','Weight','Team','Birth_Place','Birthdate','Age'])
    df2 = df1['Experience'].str.replace('R','0')
    df1['Experience'] = df2
    return df1
load_and_clean('../../data/raw/players_stats.csv')
```

## Task 3

```python
def load_and_process_data(url_or_path_to_csv_file):

    df1 = (pd.read_csv(url_or_path_to_csv_file)
          .drop(columns = ['BMI','Weight','Team','Birth_Place','Birthdate','Age'])
          .dropna()
          .replace('R', '0')
         )
    return df1
load_and_process_data('../../data/raw/players_stats.csv')
```

```python
tvp = (df1[['Name', 'PTS', 'Experience']]
       .dropna() 
      )
tvp
#Taking a DataFrame of only name, points, and time in league (experience)
```

```python
grp = (tvp.groupby('Experience')
         .mean(['PTS'])
         .reset_index()
         .sort_values(by=['PTS'], ascending = True)
        )
grp
#Each level of experience grouped and averaged PTS in season
```

```python
plt.figure(figsize=(15, 7))
sns.scatterplot(data=grp,
                x='Experience',
                y='PTS'
                )
plt.xlabel('Experience (Years in NBA)')
plt.ylabel('Points in Season')
plt.title('Points Versus Years of Experience in NBA')
```

```python
plt.figure(figsize=(15, 7))
sns.violinplot(data = tvp,
              x = 'Experience',
              y = 'PTS'
              )
#Distribution of each experience level PTS
```

```python
top = (tvp.sort_values(by='PTS', ascending = False)
          .head(5))
top
#experience of top 5 point getters
```

```python
plt.figure(figsize=(15, 7))
sns.set_context("talk")
sns.scatterplot(data = top,
               x = 'Name',
               y = 'PTS',
               size = 'Experience',
               hue = 'Experience')
```
